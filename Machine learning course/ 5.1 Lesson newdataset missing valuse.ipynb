```python
# Datasetda qaysi qatorlarida null qiymatlar bo'lsa o'sha  qatordagi barcha qiymatlarni qatori bilan tashlab yuborish

df.dropna(inplace=True)
```

```python
# datasetni null qiymatlari bor biron bir ustunini butunlay tashlab yuborish

df.drop("Name",axis=1,inplace=True)
```

```python
df.head(10)
```

```python
# Datasetdagi biron ustundagi null qiymatlarni int qilib mean(o'rtacha qiymat bilan to'ldirish

df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Biron bir ustundagi srt typle null qiymatlarini joyini to'ldirish

df.fillna(df["Name"].mode()[0],inplace=True)
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
#  Datasetdagi barcha null qiymatlari bor joylarni int va str qilib mean va mode qiymatlar bilan to'ldirish uchun umumiy code

missing_col=df.columns[df.isnull().sum()>0]
for col in missing_col:
    if df[col].dtype=="object":
        df[col].fillna(df[col].mode()[0],inplace=True)
    else:
        df[col].fillna(df[col].mean(),inplace=True)    
        
```
