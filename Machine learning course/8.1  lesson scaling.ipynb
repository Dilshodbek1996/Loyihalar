```python
# Null qiymatlari bor ustunlarni to'ldirish umumiy codi

missing_col=df.columns[df.isnull().sum()>0]

for col in missing_col:
    if df[col].dtype=="object":
        df[col].fillna(df[col].mode()[0],inplace=True)
    else:
        df[col].fillna(df[col].mean(),inplace=True)
```

# Frequency va Target (mean) encoding

```python
df.head(3)
```

```python
df.info()
```

```python
# Categorical qiymatli ustunlarni ekranga chiqarish
categorical_col=df.select_dtypes(include=["object","category"]).columns.tolist()
categorical_col
```

```python
#Frequency encoding

df[categorical_col]=df[categorical_col].apply(lambda col:col.map(col.value_counts()))
```

```python
df.head()
```

```python
df.info()
```

```python
df.info()
```

```python
# Categorical qiymatli ustunlarni ekranga chiqarish
categorical_col=df.select_dtypes(include=["object","category"]).columns.tolist()
categorical_col
```

```python
# Target(mean) encoding

df[categorical_col]=df[categorical_col].apply(lambda col:col.map(df.groupby(col)["Price"].mean()))
```
