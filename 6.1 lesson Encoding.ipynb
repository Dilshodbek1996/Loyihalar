```python
missing_col=df.columns[df.isnull().sum()>0]
for col in missing_col:
    if df[col].dtypes=="object":
        df[col].fillna(df[col].mode()[0],inplace=True)
    else:
        df[col].fillna(df[col].mean(),inplace=True)
```
